# CLEAN THE DATA 

```python
def clean_strike_rate(value):
    value = str(value)
    parts = value.split(".")
    if len(parts)>2:
        value = parts[0] + "." + "".join(parts[1:])
        return pd.to_numeric(value,errors="coerce")
df["strike_rate"]=df["strike_rate"].apply(clean_strike_rate)
df["strike_rate"].head
```

```python
df["strike_rate"].tail(5)
```

# FIX PERCENTAGES %

```python
df["average_percentage"]=df["percentage"].astype(str).str.replace(",",".").str.replace("%","").astype(float)
```
